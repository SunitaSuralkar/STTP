## LAB1 :**To understand functional programming concepts in Spark using RDD transformations and actions.**

**Concepts Used**

map()

filter()

reduce()

```python
from pyspark import SparkContext

sc = SparkContext("local", "FunctionalProgrammingLab")
```

```python
# Create RDD
numbers = sc.parallelize([1, 2, 3, 4, 5, 6])
```

```python
# Functional transformations
squares = numbers.map(lambda x: x * x)
even_numbers = numbers.filter(lambda x: x % 2 == 0)
```

```python
# Action
sum_result = numbers.reduce(lambda a, b: a + b)
```

```python
print("Squares:", squares.collect())
print("Even Numbers:", even_numbers.collect())
print("Sum:", sum_result)

sc.stop()
```
